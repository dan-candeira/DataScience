```
# https://stackoverflow.com/questions/23116330/mongodb-select-count-group-by
state_sight = list(collection.aggregate([
    {"$group" : {'Views':{'$sum':1}, '_id':"$State"}}, {'$sort': {'Views': -1}}
]))
```
